```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the datasets and read the CSV files into Pandas DataFrames
high_popularity_file = 'high_popularity_spotify_data.csv'
low_popularity_file = 'low_popularity_spotify_data.csv'

high_popularity_data = pd.read_csv(high_popularity_file)
low_popularity_data = pd.read_csv(low_popularity_file)

# First few rows of each dataset
print("High Popularity Dataset:")
display(high_popularity_data.head())
print("\nLow Popularity Dataset:")
display(low_popularity_data.head())

# Summary statistics for numerical columns
print("\nHigh Popularity Dataset Summary Statistics:")
display(high_popularity_data.describe())
print("\nLow Popularity Dataset Summary Statistics:")
display(low_popularity_data.describe())

# Basic info on both datasets
print("\nHigh Popularity Dataset Info:")
display(high_popularity_data.info())
print("\nLow Popularity Dataset Info:")
display(low_popularity_data.info())
```

```python
## Cleaning the Data ##

# Check for missing values
print("\nMissing Values in High Popularity Dataset:")
display(high_popularity_data.isnull().sum())
print("\nMissing Values in Low Popularity Dataset:")
display(low_popularity_data.isnull().sum())

# Display rows with missing values for High Popularity Dataset
print("Row with Missing Values in High Popularity Dataset:")
display(high_popularity_data[high_popularity_data.isnull().any(axis=1)])

# Drop rows with missing values for High Popularity Dataset
high_popularity_data_cleaned = high_popularity_data.dropna()
print(f"\nAfter cleaning, high popularity dataset now has {len(high_popularity_data_cleaned)} rows (from {len(high_popularity_data)}).")

# Display rows with missing values for Low Popularity Dataset
print("\nRow with Missing Values in Low Popularity Dataset:")
display(low_popularity_data[low_popularity_data.isnull().any(axis=1)])

# Drop rows with missing values for Low Popularity Dataset
low_popularity_data_cleaned = low_popularity_data.dropna()
print(f"\nAfter cleaning, low popularity dataset now has {len(low_popularity_data_cleaned)} rows (from {len(low_popularity_data)}).")
```

```python
# Initial visualization: Distribution of Spotify Popularity Scores

plt.figure(figsize=(10, 5))
plt.hist(high_popularity_data_cleaned['track_popularity'], bins=20, alpha=0.7, label='High Popularity', color='green')
plt.hist(low_popularity_data_cleaned['track_popularity'], bins=20, alpha=0.7, label='Low Popularity', color='red')
plt.title('Distribution of Spotify Popularity Scores')
plt.xlabel('Popularity Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import seaborn as sns

audio_features = ['energy', 'danceability', 'tempo', 'loudness', 'valence', 'speechiness', 
                  'instrumentalness', 'acousticness', 'liveness', 'key', 'mode', 'duration_ms']

# Add dataset type columns to distinguish high and low popularity datasetZ
high_popularity_data_cleaned = high_popularity_data_cleaned.copy()
low_popularity_data_cleaned = low_popularity_data_cleaned.copy()

high_popularity_data_cleaned.loc[:, 'dataset_type'] = 'High'
low_popularity_data_cleaned.loc[:, 'dataset_type'] = 'Low'

# Combine the datasets
combined_data = pd.concat([high_popularity_data_cleaned, low_popularity_data_cleaned], ignore_index=True)

# 1. Explore correlation between track_popularity and audio features
plt.figure(figsize=(10, 8))
numeric_data_combined = combined_data.select_dtypes(include=['float64', 'int64'])
correlation_matrix_combined = numeric_data_combined.corr()
sns.heatmap(correlation_matrix_combined, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title('Heat Map of Numerical Features (Combined Dataset)')
plt.show()


# 2. Distributions of audio features: Overlaid histograms for high and low popularity datasets
plt.figure(figsize=(15, 10))
for i, feature in enumerate(audio_features, 1):
    plt.subplot(4, 3, i)
    sns.histplot(data=combined_data, x=feature, hue='dataset_type', kde=True, palette={'High': 'blue', 'Low': 'orange'})
    plt.title(f'Distribution of {feature} (High & Low Popularity)')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 3. Track_popularity vs audio features: Overlaid scatter plots for high and low popularity datasets
plt.figure(figsize=(15, 10))
for i, feature in enumerate(audio_features, 1):
    plt.subplot(4, 3, i)
    sns.scatterplot(data=combined_data, x=feature, y='track_popularity', hue='dataset_type', alpha=0.6, palette={'High': 'blue', 'Low': 'orange'})
    plt.title(f'{feature} vs Track Popularity (High & Low Popularity)')
    plt.xlabel(feature)
    plt.ylabel('Track Popularity')
plt.tight_layout()
plt.show()

# Additional exploration of playlist genre 
plt.figure(figsize=(28, 6))
sns.boxplot(x='playlist_genre', y='track_popularity', data=combined_data)
plt.title('Track Popularity by Playlist Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Keeping only relevant features to better focus on the relationship to track_popularity
# I am going to drop tempo, mode, key and duration, even though they are audio features, because they not relevant and logically shouldn't correlate with track_popularity (and don't seem to)
# I am dropping liveness and speechiness because through trial and error, they limited the seperability of clusters later on, especially when looking at clusters' avg/median track_popularity
correlated_columns = ['energy', 'danceability', 'valence', 'loudness', 
                      'instrumentalness', 'acousticness']  

filtered_data = combined_data[correlated_columns]

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(filtered_data)

# Fit PCA to the data
pca = PCA()
pca.fit(scaled_data)

# Plot the cumulative explained variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), 
         pca.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.axhline(y=0.9, color='red', linestyle='-')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Applying PCA
n_components = 4  # Based on +90% Cumulative Explained Varience
pca = PCA(n_components=n_components)
pca_result = pca.fit_transform(scaled_data)

for i in range(n_components):
    combined_data[f'PCA_Component_{i + 1}'] = pca_result[:, i] # Add principal components to the dataset

# Calculating the explained variance of components
explained_variance = pca.explained_variance_ratio_
print(f"Explained Variance Ratios: {explained_variance}")
print(f"Total Variance Explained by {n_components} Components: {sum(explained_variance):.2f}")

# Visualize PCA results 
# (Scatter plot of first 2 components)
sns.scatterplot(data=combined_data, x='PCA_Component_1', y='PCA_Component_2', hue='dataset_type', alpha=0.6, palette={'High': 'blue', 'Low': 'orange'}, s=10)
plt.title('PCA Scatter Plot (First 2 Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()

# 3D Scatter plot for the first three components
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(combined_data['PCA_Component_1'], 
                     combined_data['PCA_Component_2'], 
                     combined_data['PCA_Component_3'], 
                     c=combined_data['track_popularity'], 
                     cmap='viridis', 
                     alpha=0.9,
                     s=10
                    )
fig.colorbar(scatter, ax=ax, label='Track Popularity')

ax.set_title('3D PCA Scatter Plot (First 3 Components)')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

plt.show()
```

```python
from sklearn.cluster import KMeans

## Performing Elbow Method to determine number of clusters
inertia = [] # inertia (sum of squared distances)

# Run KMeans for each number of clusters
for n in range(1,11):
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(pca_result)
    inertia.append(kmeans.inertia_)  # Store the inertia (sum of squared distances)

# Plot the inertia values
plt.figure(figsize=(6, 5))
plt.plot(range(1,11), inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(range(1,11))
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Performing KMeans clustering on PCA-transformed data
n_clusters = 4 # Based on Elbow Method
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
combined_data['Cluster_Label'] = kmeans.fit_predict(pca_result)

# Visualize clusters in the 3D PCA space
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    combined_data['PCA_Component_1'], 
    combined_data['PCA_Component_2'], 
    combined_data['PCA_Component_3'], 
    c=combined_data['Cluster_Label'], 
    cmap='viridis', 
    alpha=0.5
)
legend = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend)
plt.title("KMeans Clusters in PCA Space")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
ax.set_zlabel("PCA Component 3")
plt.show()

# Analyze clusters
cluster_summary = combined_data.groupby('Cluster_Label')['track_popularity'].describe()
print(cluster_summary)

# Evaluate silhouette score
silhouette_avg = silhouette_score(pca_result, combined_data['Cluster_Label'])
print(f"Silhouette Score: {silhouette_avg:.2f}")

## Investigating Popularity of Clusters
# Boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(data=combined_data, x='Cluster_Label', y='track_popularity')
plt.title('Track Popularity by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Track Popularity')
plt.ylim(0, 100)
plt.show()

# Violin plot
plt.figure(figsize=(10, 5))
sns.violinplot(data=combined_data, x='Cluster_Label', y='track_popularity', inner="quartile")
plt.title('Track Popularity by Cluster (Violin Plot)')
plt.xlabel('Cluster')
plt.ylabel('Track Popularity')
plt.ylim(0, 100)
plt.show()

## Investigating other numerial features of clusters
numerical_columns = combined_data.select_dtypes(include=['float64', 'int64']).columns.tolist()
cluster_feature_means = combined_data.groupby('Cluster_Label')[numerical_columns].mean() # Calculate mean feature values for each cluster
cluster_feature_medians = combined_data.groupby('Cluster_Label')[numerical_columns].median() # Calculate median feature values for each cluster

print("Cluster Feature Means")
display(cluster_feature_means)
print("Cluster Feature Medians")
display(cluster_feature_medians)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors

# Using a K-distance plot to find the optimal eps
# Starting with n_neighbors of 48 based off a starting point of roughly 1% of the dataset
nearest_neighbors = NearestNeighbors(n_neighbors=48)
nearest_neighbors.fit(pca_result)
distances, indices = nearest_neighbors.kneighbors(pca_result)

# Finding th elbow point
distances = np.sort(distances[:, -1])  # Take the distance to the 48th nearest neighbor
plt.figure(figsize=(8, 5))
plt.plot(distances)
plt.title("k-Distance Plot (48th Nearest Neighbor)")
plt.xlabel("Data Points Sorted by Distance")
plt.ylabel("48th Nearest Neighbor Distance")

elbow_point = 0.9
plt.axhline(y=elbow_point, color='r', linestyle='--', label='Elbow Point')
plt.legend()
plt.show()
```

```python
# Applying DBSCAN starting with the chosen eps and working down to seperate clusters
eps = 0.7
min_samples = 48
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
cluster_labels = dbscan.fit_predict(pca_result)

# Add DBSCAN labels to the dataset
combined_data['DBSCAN_Cluster'] = cluster_labels

# Visualize DBSCAN clusters in the PCA space
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    combined_data['PCA_Component_1'], 
    combined_data['PCA_Component_2'], 
    combined_data['PCA_Component_3'], 
    c=combined_data['DBSCAN_Cluster'], 
    cmap='viridis', 
    alpha=0.7
)
legend = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend)
plt.title("DBSCAN Clusters in PCA Space")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
ax.set_zlabel("PCA Component 3")
plt.show()

# noise points
noise_points = combined_data[combined_data['DBSCAN_Cluster'] == -1]
print(f"Number of Noise Points: {len(noise_points)}")

print(combined_data['DBSCAN_Cluster'].value_counts())  # Count points in each cluster
print(combined_data['DBSCAN_Cluster'].unique())  # List unique cluster labels (including noise)

# Filter out noise points (-1) from DBSCAN clusters
filtered_data = combined_data[combined_data['DBSCAN_Cluster'] != -1]  # Exclude noise
filtered_labels = filtered_data['DBSCAN_Cluster'].values  # Valid cluster labels

# Use the PCA-transformed data for Silhouette Score
pca_features = filtered_data[['PCA_Component_1', 'PCA_Component_2', 'PCA_Component_3', 'PCA_Component_4']]

# Ensure there are at least 2 clusters for valid silhouette score calculation
if len(set(filtered_labels)) > 1:
    sil_score_pca = silhouette_score(pca_features, filtered_labels)
    print(f"Silhouette Score (PCA Space): {sil_score_pca:.4f}")
else:
    print("Silhouette Score cannot be calculated as there is only one valid cluster.")


## Investigating feature values of clusters
numeric_data = combined_data.select_dtypes(include=['float64', 'int64'])

# Group by DBSCAN_Cluster and calculate median and std
cluster_summary = numeric_data.groupby(combined_data['DBSCAN_Cluster']).median()
cluster_std = numeric_data.groupby(combined_data['DBSCAN_Cluster']).std()

print("Cluster Medians:")
display(cluster_summary)

print("\nCluster Standard Deviations:")
display(cluster_std)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Extract points from the largest cluster (Cluster 0)
large_cluster_data = combined_data[combined_data['DBSCAN_Cluster'] == 0].copy()
pca_features_large_cluster = large_cluster_data[['PCA_Component_1', 'PCA_Component_2', 'PCA_Component_3', 'PCA_Component_4']]

# Perform hierarchical clustering
linkage_matrix = linkage(pca_features_large_cluster, method='complete')

# Plot the dendrogram
plt.figure(figsize=(12, 8))
dendrogram(linkage_matrix, truncate_mode='level', p=5)  # Adjust 'p' to control depth of dendrogram
plt.title("Hierarchical Clustering Dendrogram (Cluster 0)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.axhline(y=4.5, color='red', linestyle='-')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

# Hierarchical Clustering
num_subclusters = 4
subcluster_labels = fcluster(linkage_matrix, num_subclusters, criterion='maxclust')

# Add subcluster labels to the dataset
large_cluster_data['Subcluster_Label'] = subcluster_labels

# Visualizing subclusters in 2D PCA space
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    large_cluster_data['PCA_Component_1'],
    large_cluster_data['PCA_Component_2'],
    c=large_cluster_data['Subcluster_Label'],
    cmap='viridis',
    alpha=0.7
)
plt.title("Subclusters in PCA Space (Cluster 0)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(scatter, label="Subcluster Label")
plt.show()

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    large_cluster_data['PCA_Component_1'], 
    large_cluster_data['PCA_Component_2'], 
    large_cluster_data['PCA_Component_3'], 
    c=large_cluster_data['Subcluster_Label'], 
    cmap='viridis', 
    alpha=0.7
)
legend = ax.legend(*scatter.legend_elements(), title="Subclusters")
ax.add_artist(legend)
plt.title("Hierarchical Subclusters in PCA Space")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
ax.set_zlabel("PCA Component 3")
plt.show()

## Summary Statistics of Subclusters

# Drop unnecessary columns from numerical data
columns_to_remove = ['PCA_Component_1', 'PCA_Component_2', 'PCA_Component_3', 'PCA_Component_4', 'PCA_Component_5', 'Cluster_Label']
filtered_numeric_data = large_cluster_data[[col for col in numerical_columns if col not in columns_to_remove]]

# Calculate medians and standard deviations for each subcluster
cluster_medians = filtered_numeric_data.groupby(large_cluster_data['Subcluster_Label']).median()
cluster_means = filtered_numeric_data.groupby(large_cluster_data['Subcluster_Label']).mean()
cluster_std = filtered_numeric_data.groupby(large_cluster_data['Subcluster_Label']).std()

# Calculate the count for each subcluster
cluster_counts = filtered_numeric_data.groupby(large_cluster_data['Subcluster_Label']).size()
cluster_medians['Count'] = cluster_counts

# Display Medians, Means and Std
print("Subcluster Medians:")
display(cluster_medians)
print("Subcluster Means:")
display(cluster_means)
print("\nSubcluster Standard Deviations:")
display(cluster_std)

# Violin plot (Track Popularity)
plt.figure(figsize=(10, 5))
sns.violinplot(data=large_cluster_data, x='Subcluster_Label', y='track_popularity', inner="quartile")
plt.title('Track Popularity by Subluster (Violin Plot)')
plt.xlabel('Subluster')
plt.ylabel('Track Popularity')
plt.ylim(0, 100)
plt.show()


## Comparing the subcluster with the high_popularity_data and combined_data to see significant differences

# Remove unnecessary columns from combined_data and calculate medians
columns_to_remove_combined = ['PCA_Component_1', 'PCA_Component_2', 'PCA_Component_3', 
                              'PCA_Component_4', 'Cluster_Label', 'DBSCAN_Cluster']
filtered_combined_columns = [col for col in numerical_columns if col not in columns_to_remove_combined] 
numeric_combined_data = combined_data[filtered_combined_columns]
combined_data_medians = numeric_combined_data.median() # Calculate medians for combined_data

# Remove unnecessary columns from high_popularity_data_cleaned and calculate medians
filtered_high_popularity_columns = [col for col in numerical_columns if col in high_popularity_data_cleaned.columns] # Filter the numerical columns for high_popularity_data_cleaned
numeric_high_popularity_data = high_popularity_data_cleaned[filtered_high_popularity_columns]
high_popularity_medians = numeric_high_popularity_data.median() # Calculate medians for high_popularity_data

prepared_cluster_medians = cluster_medians.drop(columns=['DBSCAN_Cluster'], errors='ignore') # Prepare cluster_medians by removing DBSCAN_Cluster column

# Align medians to the same structure
aligned_combined_data_medians = combined_data_medians.reindex(prepared_cluster_medians.columns, fill_value=np.nan)
aligned_high_popularity_medians = high_popularity_medians.reindex(prepared_cluster_medians.columns, fill_value=np.nan)

# Calculate differences
combined_data_differences = aligned_combined_data_medians - prepared_cluster_medians # Differences for combined_data
high_popularity_differences = aligned_high_popularity_medians - prepared_cluster_medians # Differences for high_popularity_data_cleaned

# Step 5: Combine results for display
# concatenate the results into a single DataFrame for easier comparison
comparison_results = pd.concat({
    "Cluster_Medians": prepared_cluster_medians,
    "Combined_Data_Differences": combined_data_differences,
    "High_Popularity_Differences": high_popularity_differences
}, axis=0)

# Display the results
print("Comparison Results:")
display(comparison_results)
```

```python
from scipy.stats import median_abs_deviation

combined_data_mad = numeric_combined_data.apply(median_abs_deviation, scale='normal', axis=0)
high_pop_mad = numeric_high_popularity_data.apply(median_abs_deviation, scale='normal', axis=0)

aligned_combined_data_medians = aligned_combined_data_medians.drop('Count', errors='ignore')
aligned_high_popularity_medians = aligned_high_popularity_medians.drop('Count', errors='ignore')

# Standardize the Cluster_Medians
# Prepare Cluster_Medians (make sure it doesn't contain count)
numeric_cluster_medians = prepared_cluster_medians.drop(columns=['Count'], errors='ignore')

# Calculate Z-scores (standardized values)
z_scores = (numeric_cluster_medians - aligned_combined_data_medians) / combined_data_mad
z_scores = z_scores.fillna(0)
z_scores_popular = (numeric_cluster_medians - aligned_high_popularity_medians) / high_pop_mad
z_scores_popular = z_scores_popular.fillna(0)

## Heatmaps of features in standard deviations
plt.figure(figsize=(12, 8))
sns.heatmap(z_scores.T, annot=True, cmap='coolwarm', center=0, cbar_kws={'label': 'MAD from Median'})
plt.title('Subcluster Medians (Standardized by Combined Data)')
plt.ylabel('Features')
plt.xlabel('Subcluster Label')
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(z_scores_popular.T, annot=True, cmap='coolwarm', center=0, cbar_kws={'label': 'MAD from Median'})
plt.title('Subcluster Medians (Standardized by High_Popularity Data)')
plt.ylabel('Features')
plt.xlabel('Subcluster Label')
plt.show()
```

```python
# Merge the dataframes based on matching song names and artists
merged_data = pd.merge(
    combined_data, 
    additional_data, 
    left_on=['track_name', 'track_album_name', 'track_artist'], 
    right_on=['Track', 'Album Name', 'Artist'], 
    how='inner',
    suffixes=('', '_dup')  # Keep original names in `large_cluster_data` and rename duplicates
)
merged_data = merged_data.loc[:, ~merged_data.columns.str.endswith('_dup')] # Drop duplicate columns (with the `_dup` suffix)

merged_large_cluster = pd.merge(
    large_cluster_data,
    additional_data,
    left_on=['track_name', 'track_album_name', 'track_artist'],
    right_on=['Track', 'Album Name', 'Artist'],
    how='inner',
    suffixes=('', '_dup')  # Keep original names in `large_cluster_data` and rename duplicates
)
merged_large_cluster = merged_large_cluster.loc[:, ~merged_large_cluster.columns.str.endswith('_dup')] # Drop duplicate columns


## Evaluating each Clustering Step

fig, axes = plt.subplots(2, 3, figsize=(20, 6), gridspec_kw={'height_ratios': [3, 1]}) # Figure with 1 row and 3 columns for bar plots

# --- KMeans Graph ---
counts_merged = merged_data['Cluster_Label'].value_counts().sort_index()
counts_combined = combined_data['Cluster_Label'].value_counts().sort_index()
all_clusters = sorted(set(counts_merged.index).union(set(counts_combined.index)))
counts_merged = counts_merged.reindex(all_clusters, fill_value=0)
counts_combined = counts_combined.reindex(all_clusters, fill_value=0)
x = np.arange(len(all_clusters))  # The x locations for the clusters
width = 0.32

# Bar graph
axes[0, 0].bar(x - width/2, counts_merged, width, label='Merged Data', color='blue', alpha=0.7)
axes[0, 0].bar(x + width/2, counts_combined, width, label='Combined Data', color='green', alpha=0.7)
axes[0, 0].set_xlabel('Cluster Label')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_title('KMeans Cluster Counts vs Combined Data')
axes[0, 0].set_xticks(x)
axes[0, 0].set_xticklabels(all_clusters)
axes[0, 0].legend()

# Bottom text (percentage breakdown)
axes[1, 0].axis('off')  # Hide the axis
percentage_breakdown_1 = counts_merged / counts_combined * 100
percentage_breakdown_1 = percentage_breakdown_1.fillna(0)  # Handle division by zero if any cluster has no counts
formatted_percentage_breakdown_1 = percentage_breakdown_1.apply(lambda x: f"{x:.2f}%")
percentage_text_1 = "\n".join([f"{k}: {v}" for k, v in formatted_percentage_breakdown_1.items()])
axes[1, 0].text(0.5, 0.5, percentage_text_1, fontsize=10, ha='center', va='center', transform=axes[1, 0].transAxes)


# --- DBSCAN Graph ---
counts_merged_DBSCAN = merged_data['DBSCAN_Cluster'].value_counts().sort_index()
counts_combined_DBSCAN = combined_data['DBSCAN_Cluster'].value_counts().sort_index()
all_clusters = sorted(set(counts_merged_DBSCAN.index).union(set(counts_combined_DBSCAN.index))) # Align indexes
counts_merged_DBSCAN = counts_merged_DBSCAN.reindex(all_clusters, fill_value=0)
counts_combined_DBSCAN = counts_combined_DBSCAN.reindex(all_clusters, fill_value=0)
x = np.arange(len(all_clusters))  # The x locations for the clusters

# Bar graph
axes[0, 1].bar(x - width/2, counts_merged_DBSCAN, width, label='Merged Data', color='blue', alpha=0.7)
axes[0, 1].bar(x + width/2, counts_combined_DBSCAN, width, label='Combined Data', color='green', alpha=0.7)
axes[0, 1].set_xlabel('DBSCAN Cluster')
axes[0, 1].set_ylabel('Count')
axes[0, 1].set_title('DBSCAN Cluster Counts vs Combined Data')
axes[0, 1].set_xticks(x)
axes[0, 1].set_xticklabels(all_clusters)
axes[0, 1].legend()

# Bottom text (percentage breakdown)
axes[1, 1].axis('off')
percentage_breakdown_2 = counts_merged_DBSCAN / counts_combined_DBSCAN * 100
percentage_breakdown_2 = percentage_breakdown_2.fillna(0)  # Handle division by zero if any cluster has no counts
formatted_percentage_breakdown_2 = percentage_breakdown_2.apply(lambda x: f"{x:.2f}%")
percentage_text_2 = "\n".join([f"{k}: {v}" for k, v in formatted_percentage_breakdown_2.items()])
axes[1, 1].text(0.5, 0.5, percentage_text_2, fontsize=10, ha='center', va='center', transform=axes[1, 1].transAxes)

# --- Subclusters Graph ---
counts_merged_SC = merged_large_cluster['Subcluster_Label'].value_counts().sort_index()
counts_large_cluster = large_cluster_data['Subcluster_Label'].value_counts().sort_index()
all_subclusters = sorted(set(counts_merged_SC.index).union(set(counts_large_cluster.index))) # Line up indexes
counts_merged_SC = counts_merged_SC.reindex(all_subclusters, fill_value=0)
counts_large_cluster = counts_large_cluster.reindex(all_subclusters, fill_value=0)
x = np.arange(len(all_subclusters))  # The x locations for the subclusters

# Bar graph
axes[0, 2].bar(x - width/2, counts_merged_SC, width, label='Merged Large Cluster', color='blue', alpha=0.7)
axes[0, 2].bar(x + width/2, counts_large_cluster, width, label='Large Cluster Data', color='orange', alpha=0.7)
axes[0, 2].set_xlabel('Subcluster Label')
axes[0, 2].set_ylabel('Count')
axes[0, 2].set_title('Subcluster Counts vs DBSCAN_Cluster')
axes[0, 2].set_xticks(x)
axes[0, 2].set_xticklabels(all_subclusters)
axes[0, 2].legend()

# Bottom text (percentage breakdown)
axes[1, 2].axis('off') 
percentage_breakdown_3 = counts_merged_SC / counts_large_cluster * 100
percentage_breakdown_3 = percentage_breakdown_3.fillna(0)  # Handle division by zero if any cluster has no counts
formatted_percentage_breakdown_3 = percentage_breakdown_3.apply(lambda x: f"{x:.2f}%")
percentage_text_3 = "\n".join([f"{k}: {v}" for k, v in formatted_percentage_breakdown_3.items()])
axes[1, 2].text(0.5, 0.5, percentage_text_3, fontsize=10, ha='center', va='center', transform=axes[1, 2].transAxes)

plt.tight_layout()
plt.show()
```
